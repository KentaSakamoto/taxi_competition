```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv('../data/JR201604.csv',sep=',',encoding='cp932', index_col=0) #index_col=0 : unnamedを省ける
```

```python
df.head()
```

```python
df.tail()
```

```python
print(df.dtypes)
```

```python
df['StatusTime'] = pd.to_datetime(df['StatusTime'])
print(df.dtypes)
```

## 2016年4月11日から15日までの一週間を対象にそれぞれ一日ごとのデータに分割するデータセットを作成する

```python
JR20160411 = df[(df['StatusTime'] >= dt.datetime(2016,4,11,8,0,0,0)) & (df['StatusTime'] <= dt.datetime(2016,4,11,19,0,0,0))]
```

```python
JR20160411.head()
```

```python
JR20160411.tail()
```

```python
JR20160411.to_csv('../data/JR20160411.csv')
```

### 同様にして12日から15日まで作成する

```python
JR20160412 = df[(df['StatusTime'] >= dt.datetime(2016,4,12,8,0,0,0)) & (df['StatusTime'] <= dt.datetime(2016,4,12,19,0,0,0))]
JR20160412.to_csv('../data/JR20160412.csv')

JR20160413 = df[(df['StatusTime'] >= dt.datetime(2016,4,13,8,0,0,0)) & (df['StatusTime'] <= dt.datetime(2016,4,13,19,0,0,0))]
JR20160413.to_csv('../data/JR20160413.csv')

JR20160414 = df[(df['StatusTime'] >= dt.datetime(2016,4,14,8,0,0,0)) & (df['StatusTime'] <= dt.datetime(2016,4,14,19,0,0,0))]
JR20160414.to_csv('../data/JR20160414.csv')

JR20160415 = df[(df['StatusTime'] >= dt.datetime(2016,4,15,8,0,0,0)) & (df['StatusTime'] <= dt.datetime(2016,4,15,19,0,0,0))]
JR20160415.to_csv('../data/JR20160415.csv')
```
